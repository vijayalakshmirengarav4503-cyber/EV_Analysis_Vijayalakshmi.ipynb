```python
# Filter data to include Model Year 2010 and later for a clearer trend visualization
df_filtered = df[df['Model Year'] >= 2010].copy()

# Group by 'Model Year' and count the number of electric vehicles
ev_adoption_trend = df_filtered.groupby('Model Year').size().reset_index(name='EV Count')

# Ensure Model Year is treated as an integer for correct sorting
ev_adoption_trend['Model Year'] = ev_adoption_trend['Model Year'].astype(int)

# Create a line graph for the adoption trend
plt.figure(figsize=(12, 6))
plt.plot(ev_adoption_trend['Model Year'], ev_adoption_trend['EV Count'], marker='o', linestyle='-', color='green')
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Number of Electric Vehicles (EVs)', fontsize=12)
plt.title('Trend of EV Adoption by Model Year (2010+)', fontsize=14)
plt.xticks(ev_adoption_trend['Model Year'].unique(), rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Print the data used for the plot
print("EV Adoption Trend by Model Year (2010+):\n")
print(ev_adoption_trend.to_markdown(index=False))
```

```python
# Filter data: Remove entries where MSRP or Electric Range are reported as 0
# A Base MSRP of 0 usually indicates missing data, and Electric Range of 0 is ambiguous for trend analysis.
df_filtered = df[(df['Base MSRP'] > 0) & (df['Electric Range'] > 0)].copy()

# Ensure the columns are numeric (they should be, but good practice)
df_filtered['Base MSRP'] = pd.to_numeric(df_filtered['Base MSRP'])
df_filtered['Electric Range'] = pd.to_numeric(df_filtered['Electric Range'])

# Calculate the count of valid data points
valid_count = len(df_filtered)

# Create the scatter plot
plt.figure(figsize=(10, 6))
# Using a small alpha to show density in areas where points overlap
plt.scatter(df_filtered['Electric Range'], df_filtered['Base MSRP'], alpha=0.3, color='#0072B2', s=10)

plt.xlabel('Electric Range (Miles)', fontsize=12)
plt.ylabel('Base MSRP ($)', fontsize=12)
plt.title(f'Pricing Trend: Electric Range vs. Base MSRP (n={valid_count} vehicles)', fontsize=14)
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Set reasonable limits for better visualization, based on general EV ranges/prices
plt.xlim(0, df_filtered['Electric Range'].max() * 1.05)
plt.ylim(0, df_filtered['Base MSRP'].max() * 1.05)

plt.tight_layout()
plt.savefig('range_vs_msrp_scatter_plot.png')

# Print summary statistics of the filtered data
print(f"Total valid data points for plot (MSRP > 0 and Range > 0): {valid_count}\n")
print("Summary Statistics for Plot Variables:")
print(df_filtered[['Electric Range', 'Base MSRP']].describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Group by 'Clean Alternative Fuel Vehicle (CAFV) Eligibility' and count the records
eligibility_counts = df.groupby('Clean Alternative Fuel Vehicle (CAFV) Eligibility').size().reset_index(name='Count')

# Calculate the percentage for each category
total_count = eligibility_counts['Count'].sum()
eligibility_counts['Percentage'] = (eligibility_counts['Count'] / total_count) * 100
eligibility_counts['Percentage'] = eligibility_counts['Percentage'].round(2)

# Sort by count for better visualization order (optional, but good practice for clarity)
eligibility_counts = eligibility_counts.sort_values(by='Count', ascending=False)

# Prepare data for the pie chart
labels = eligibility_counts['Clean Alternative Fuel Vehicle (CAFV) Eligibility']
sizes = eligibility_counts['Count']
# Use autopct to display the percentage on the wedges
def func(pct, allvals):
    absolute = int(round(pct/100.*allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# Define colors for the wedges
# Colors chosen to contrast eligibility statuses
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(sizes,
        labels=labels,
        autopct=lambda pct: func(pct, total_count),
        startangle=140,
        colors=colors[:len(labels)],
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True})

plt.title('Proportion of EVs by CAFV Eligibility Status', fontsize=16, pad=20)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

# Save the plot
plt.savefig('cafv_eligibility_pie_chart.png')

# Print the data used for the plot
print("EV Proportion by CAFV Eligibility:\n")
print(eligibility_counts.to_markdown(index=False, numalign="left", stralign="left"))
```
